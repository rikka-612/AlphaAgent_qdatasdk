```python
from sqlalchemy import and_
from sqlalchemy.orm import Query

from qdatasdk import Security, run_query, EquFreeShares, EquShareChange


# 查询自由流通股本
fields = [EquFreeShares.ID, EquFreeShares.PARTY_ID,
          EquFreeShares.PUBLISH_DATE, EquFreeShares.CHANGE_DATE,
          EquFreeShares.TICKER_SYMBOL,
          EquFreeShares.FREE_SHARES]
q = Query(fields).filter(and_(EquFreeShares.PARTY_ID == Security.PARTY_ID,
                              Security.TICKER_SYMBOL == "000001",
                              EquFreeShares.PUBLISH_DATE >= "1900-01-01")).order_by(
    EquFreeShares.PUBLISH_DATE.desc()).limit(5)

df = run_query(q)
df
```

```python
# 查询总股本
fields = [EquShareChange.ID, EquShareChange.PARTY_ID,
          EquShareChange.PUBLISH_DATE, EquShareChange.CHANGE_DATE,
          EquShareChange.TOTAL_SHARES]
q = Query(fields).filter(and_(EquShareChange.PARTY_ID == Security.PARTY_ID,
                              Security.TICKER_SYMBOL == "000001",
                              EquShareChange.PUBLISH_DATE >= "1900-01-01")).order_by(
    EquShareChange.PUBLISH_DATE.desc()).limit(5)
df = run_query(q)
df
```
